```python
import numpy as np
from sklearn.model_selection import train_test_split
import pandas as pd
```

```python
df = pd.read_csv('dataset.csv')
df
```

```python
train, test = train_test_split(df, test_size = 0.2, random_state = 1)
```

```python
uniqueValues = train['phase'].unique()
print(uniqueValues)
```

```python
grouped = train.groupby(train.phase)
```

```python
SS_grp      = grouped.get_group("SS")
AM_grp      = grouped.get_group("AM")
IM_grp      = grouped.get_group("IM")
```

```python
def gen(vec,diffn,diffr,mixingh,mixinge):
    distance=[]
    distance_1 = []
    
    distance.append(((vec-SS_grp['vec'].mean())**2+(diffn-SS_grp['diffn'].mean())**2+(diffr-SS_grp['diffr'].mean())**2+(mixingh-SS_grp['mixingh'].mean())**2+(mixinge-SS_grp['mixinge'].mean())**2)**0.5)
    distance.append(((vec-AM_grp['vec'].mean())**2+(diffn-AM_grp['diffn'].mean())**2+(diffr-AM_grp['diffr'].mean())**2+(mixingh-AM_grp['mixingh'].mean())**2+(mixinge-AM_grp['mixinge'].mean())**2)**0.5)
    distance.append(((vec-IM_grp['vec'].mean())**2+(diffn-IM_grp['diffn'].mean())**2+(diffr-IM_grp['diffr'].mean())**2+(mixingh-IM_grp['mixingh'].mean())**2+(mixinge-IM_grp['mixinge'].mean())**2)**0.5)
    
    distance_1 = [x for x in distance if ~np.isnan(x)]
    
    key = min(distance_1)
    for i in range(0,3):
        if key == distance[i]:
             a = i
    print(distance_1)
    
    if (a == 0):
        a1="SS"
        return(a1)
    elif(a == 1):
        a2="AM"
        return(a2)
    elif(a == 2):
        a3="IM"
        return(a3)
```

```python
count=0
total_comp=len(test)

for index,row in test.iterrows( ):
    fav = gen(row['vec'],row['diffn'],row['diffr'],row['mixingh'],row['mixinge'])
    if(fav == row['phase']):
        count = count+1
    print(fav)
    print(row['phase'])

print(count)
print(total_comp)
print((count/total_comp)*100)
```
